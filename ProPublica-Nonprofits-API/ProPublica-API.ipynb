```python
# give number of organizations for a particular non-profit category ("ntee[id]")
# 1 = Arts, Culture, & Humanities; 2 = Education; 3 = Environment and Animals
# see documentation for others

categories = {
    1: "Arts, Culture & Humanities",
    2: "Education",
    3: "Environment and Animals",
    4: "Health",
    5: "Human Services",
    6: "International, Foreign Affairs",
    7: "Public, Societal Benefit",
    8: "Religion Related",
    9: "Mutual/Membership Benefit",
    10: "Unknown, Unclassified",
}
categories
```

```python
num_nonprofits = []

for category in categories.keys():
    num_nonprofits.append(
        requests.get(url, params={"state[id]": "TX", "ntee[id]": category}).json()[
            "total_results"
        ]
    )

tx_categories_df = pd.DataFrame(num_nonprofits, index=categories.values()).rename(
    columns={0: "num_nonprofits"}
)

tx_categories_df
```
